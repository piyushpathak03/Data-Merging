# Merge on Covid-19 Data

```python
# Import libraries
from pyforest import*
```

## Import all data from Kaggle


https://www.kaggle.com/sudalairajkumar/covid19-in-india?rvi=1

```python
# Import Age dataset
age=pd.read_csv('AgeGroupDetails.csv')
print(age.shape)
age
```

```python
# Import covid cases dataset
data=pd.read_csv('covid_19_india.csv')
print(data.shape)
data.tail()
```

```python
# Import State wise dataset
state_data=pd.read_csv('StatewiseTestingDetails.csv')
print(state_data.shape)
state_data.head()
```

### Convert date feature for better merging:- As covid data and state data both have different date format

```python
a=(state_data['Date'].map(lambda x:x.split('-')))
a.head()
```

```python
state_data.drop('Date',axis=1,inplace=True)
```

```python
state_data['Date']=a
state_data.head()
```

```python
a[0][2]+'/'+a[0][1]+'/'+a[0][0][2:4]
```

```python
def date_n(i):
    b=0
    for i in a:
        c= a[b][2]+'/'+a[b][1]+'/'+a[b][0][2:4]
        b=b+1
        return c
```

```python
state_data["Date"] = state_data.Date.apply(date_n)
```

```python
print(state_data.shape)
state_data.head()
```

```python
# Import Hospital dataset
hospital=pd.read_csv('HospitalBedsIndia.csv')
print(hospital.shape)
hospital.head()
```

```python
# Import population dataset
population=pd.read_csv('population_india_census2011.csv')
print(population.shape)
population.head()
```

# Data concatenation:-


### Merge data and state_data

```python
data.head()
```

```python
state_data.head()
```

```python
df=pd.merge(data,state_data,how='left',on=['State','Date'])
df.shape
```

```python
df[df.Date=='17/04/20'].head()
```

### Merge new data( df ) and population data

```python
df.State.unique()
```

```python
population
```

```python
df=pd.merge(df,population,how='left',on=['State'])
df.shape
```

```python
df.tail()
```

### Merge new data( df ) and hospital data

```python
hospital=hospital[:36]              #Remove all india dataset
hospital
```

```python
hospital.isnull().sum()
```

```python
df=pd.merge(df,hospital,how='left',on='State')
```

```python
df.shape
```

```python
df.head()
```
